**Q2. State and Explain the features of MongoDB.**

MongoDB is a popular NoSQL database that provides several features that make it a good choice for modern web applications.

Document-oriented:
MongoDB is a document-oriented database, which means that it stores data in a format that closely resembles the JSON data format. This makes it easy to work with data in MongoDB because it can be easily transformed into and from JSON data.

Scalability:
MongoDB is designed to be highly scalable and can be used in distributed environments to handle large amounts of data. It provides features like sharding, which allows you to split a large database into smaller, more manageable parts that can be distributed across multiple servers.

High availability:
MongoDB provides features like replica sets, which allow you to create multiple copies of your data on different servers. This provides high availability and ensures that your data is always accessible, even if one of your servers goes down.

Flexible data modeling:
MongoDB is schema-less, which means that you can store data of any structure without the need to define a schema first. This provides flexibility in data modeling and allows you to quickly adapt your data structures to changing application requirements.

Querying:
MongoDB provides a powerful and flexible querying language that allows you to perform complex queries on your data. You can use features like indexes and aggregation pipelines to optimize your queries and improve performance.

**Q7. Explain why delete_one(), delete_many(), and drop() is used.**

The delete_one() method is used to delete a single document that matches a specified filter criteria. If multiple documents match the filter, only the first matching document will be deleted.

The delete_many() method is used to delete all documents that match a specified filter criteria.

The drop() method is used to delete an entire collection from a database. This method removes all documents in the collection and also deletes the collection itself.

**Q1. What is Flask Framework? What are the advantages of Flask Framework?**

**Ans:**

Flask is a micro web framework written in Python. It is classified as a micro framework because it does not require particular tools or libraries. Flask is used for building web applications, and it follows the Model-View-Controller (MVC) architecture pattern. Flask is easy to use, flexible, and scalable, and it is suitable for building small to medium-sized web applications.

Some advantages of Flask Framework are:

Lightweight: Flask is lightweight and minimalistic, which means it has fewer dependencies and provides only the essential tools to build a web application.

Flexibility: Flask is highly flexible and allows developers to customize and modify the framework to suit their requirements.

Easy to use: Flask is easy to learn and use, which makes it an excellent choice for beginners and developers who want to build web applications quickly.

Modular design: Flask is designed to be modular, which means developers can choose the components they need and use them as required. This makes the framework scalable and efficient.

Extensible: Flask is extensible, which means developers can add new features and functionality to the framework easily.

Integration: Flask can integrate with other technologies and libraries, such as SQLalchemy, WTForms, and Jinja2, which makes it a powerful web framework.

**Q3. What is App routing in Flask? Why do we use app routes?**

**Ans**

In Flask, app routing refers to the process of mapping URLs or routes to view functions, which are Python functions that handle HTTP requests and return HTTP responses. App routing is an essential component of a Flask application, as it determines how incoming requests are processed and which view function is called to generate the corresponding response.

We use app routes in Flask for the following reasons:

Handling user requests: App routes allow developers to map URLs to view functions that handle incoming user requests. This enables developers to create dynamic web applications that can respond to user input and generate custom responses based on user actions.

Creating RESTful APIs: Flask can be used to create RESTful APIs that provide data and resources to client applications. App routes can be used to map API endpoints to view functions that generate JSON responses, which can be consumed by client applications.

SEO optimization: App routes can be used to create search engine-friendly URLs that improve the visibility and ranking of a Flask application in search engine results pages (SERPs).

Implementing authentication and authorization: App routes can be used to implement authentication and authorization functionality in a Flask application. For example, certain routes can be restricted to authenticated users only, while others can be accessible to all users.

Modular design: App routes can be used to modularize a Flask application by grouping related views and functionality together into separate blueprints. This improves the maintainability and scalability of the application by separating concerns and reducing code complexity.

**Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.**

**ANS:**

In Flask, the url_for() function is used for URL building. This function generates a URL to a specific function based on the given endpoint and other parameters.

```python
from flask import Flask, url_for
app = Flask(__name__)

@app.route("/welcome")
def test1():
    return "<h2>Welcome to ABC Corporation</h2>"

@app.route("/")
def test2():
    return "Company Name: ABC Corporation</h3> <br> Location: India </br> Contact Detail: 999-999-9999"

with app.test_request_context():
    print(url_for('test1')) 
    print(url_for('test2'))
```

### Solution 1

```python
marks = input("Enter the marks: ")                                # marks represent the percentage scored by a Student
try:
    marks = float(marks)
    if marks < 0 or marks>100:
        print('Please Enter the valid Input (from 0 to 100)')
    elif marks>90:
        print(f'Your percentage is {marks} & Grade is A')
    elif marks>80 and marks<=90:
        print(f'Your percentage is {marks} & Grade is B')
    elif marks>=60 and marks<=80:
        print(f'Your percentage is {marks} & Grade is C')
    else:
        print(f'Your percentage is {marks} & Grade is D')
        
except ValueError:
    print("Input a valid Number please")

```

### Solution 2

```python
cp = input('Enter the Cost Price of the Bike: ')    # cp represent the cost price of the bike
try:
    cp=float(cp)
    tax_percentage=0                                # tax_percentage is the percentage of tax which is to be paid.
    tax=0                                           # tax represent the amount that needs to be paid as road tax.
    if cp<0:
        print('Please Enter the valid Price')
    elif cp>100000:
        tax_percentage = 15/100
        tax = cp*tax_percentage
    elif cp>5000 and cp<=100000:
        tax_percentage = 10/100
        tax = cp*tax_percentage
    else:
        tax_percentage = 5/100
        tax = cp*tax_percentage
    
    print(f'The Cost Price of Bike is Rs.{cp}, the Tax Percentage is {tax_percentage*100}% & Road Tax is Rs.{tax}')
except ValueError:
    print('Please Enter the valid Input')
```

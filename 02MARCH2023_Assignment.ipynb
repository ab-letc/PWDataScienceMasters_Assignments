**Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.**

#### Ans:

Matplotlib is a python library which is utilised for visualization of data through various plots. These plots can be 2d or 3d depedning upon our requirement. 

**Various plots that can be done using Matplotlib.pyplot are:**

Line plot: A line plot is a type of plot that displays data points connected by straight lines. It is used to show trends over time or to compare two or more variables.

Scatter plot: A scatter plot is a type of plot that displays the relationship between two variables. It is used to show patterns or correlations between variables.

Bar plot: A bar plot is a type of plot that displays categorical data with rectangular bars. It is used to compare the values of different categories.

Histogram: A histogram is a type of plot that displays the distribution of a variable. It is used to show the frequency of occurrence of values in a dataset.

Pie chart: A pie chart is a type of plot that displays the proportion of each category in a dataset. It is used to show how the data is distributed across different categories.

```python
# Pie Chart
import matplotlib.pyplot as plt
marks = [72, 76, 95, 89,93]
subjects = ['Physics', 'Chemistry', 'English', 'Computer Science','Mathematics']
plt.title('Marks Distribution')
plt.pie(marks, labels=subjects, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','pink'], autopct='%1.1f%%')
plt.show()
```

```python
#Bar Graph
marks = [72, 76, 95, 89,93]
subjects = ['Physics', 'Chemistry', 'English', 'Computer Science','Mathematics']
colors=['#8BC34A', '#03A9F4', '#FFC107', '#E91E63', '#9C27B0']
plt.figure(figsize=(8,4))
plt.xlabel('Subjects')
plt.ylabel('Marks Obtained')
plt.title('Marks Distribution')
plt.bar(subjects,marks,color=colors)
plt.show()
```

```python
#Histogram
import numpy as np
marks = [72, 76, 95, 89,93]
plt.hist(marks,color='#009688', alpha=0.8)
plt.xlim([70,100])
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
```

```python
#Scatter Plot
x=10*np.random.rand(20)
y=10*np.random.rand(20)
plt.scatter(x,y,c='#FFC107')
plt.grid(2)
plt.xlabel('X-axis-->')
plt.ylabel('Y-axis-->')
plt.title('Scatter Plot')
plt.show()
```

```python
#Line Plot
x=10*np.random.rand(5)
y=10*np.random.rand(5)
plt.plot(x,y,c='#FFC107')
plt.grid(1)
plt.xlabel('X-axis-->')
plt.ylabel('Y-axis-->')
plt.title('Line Plot')
plt.show()
```

**Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.**

**Use the following data:**

**import numpy as np**

**For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])**

**For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])**

**For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])**

**For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])**

#### Ans:

subplot() function is used to plot multiple graphs on a single figure. It is useful when you want to compare different plots or visualize multiple data sets in one figure.

```python
import matplotlib.pyplot as plt
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# create a figure with four subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# plot each line on a different subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

plt.show()
```

**Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.**

**import numpy as np**

**company = np.array(["Apple", "Microsoft", "Google", "AMD"])**

**profit = np.array([3000, 8000, 1000, 10000])**


#### Ans:

A bar plot, also known as a bar chart or bar graph, is a type of plot that displays data using rectangular bars. The length or height of each bar represents the numerical value of the data being displayed.

Bar plots are commonly used to compare data across different categories or groups. The categories are displayed along the x-axis, while the height or length of each bar represents the value of the data being displayed along the y-axis.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
colors=['#8BC34A', '#03A9F4', '#FFC107', '#E91E63']
plt.style.use('seaborn-white')
plt.figure(figsize=(8,4))
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Profit')
plt.bar(company,profit,color=colors)
plt.show()
```

**Q5: What is a box plot? Why is it used? Using the following data plot a box plot.**

**box1 = np.random.normal(100, 10, 200)**

**box2 = np.random.normal(90, 20, 200)**

#### Ans:

A box plot, also known as a box-and-whisker plot, is a statistical visualization tool used to display the distribution of a dataset. It provides a concise summary of the data distribution by showing the median, the quartiles (Q1 and Q3), the minimum and maximum values, and any outliers.

In a box plot, a box is drawn from the first quartile to the third quartile, with a line inside representing the median value. Whiskers extend from the box to the minimum and maximum values, excluding outliers which are displayed as individual points. Outliers are typically defined as values that are more than 1.5 times the interquartile range (IQR) away from the nearest quartile.

```python
import matplotlib.pyplot as plt
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.style.use('dark_background')
plt.boxplot([box1, box2])
plt.show()
```
